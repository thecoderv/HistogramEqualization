```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("pic1color512.jpg")
s=img.shape
imgGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imwrite('grayscaleImage.png',imgGray)
plt.imshow(imgGray,cmap='gray')
plt.show()
cv2.waitKey(0)
```

```python
def Hist(image):
    H=np.zeros(shape=(256,1))
    s=image.shape
    for i in range(s[0]):
        for j in range(s[1]):
            k=image[i,j]
            H[k,0]=H[k,0]+1
    return H
```

```python
histg = Hist(imgGray)
x = histg.reshape(1,256)
y=np.array([])
y = np.append(y,x[0,0])
print("The shape of my histogram",histg.shape)
```

```python
plt.figure()
plt.plot(histg)
plt.savefig('Unequalized Histogram')
plt.show()
```

```python
for i in range(255):
    k = x[0,i+1]+y[i]
    y = np.append(y,k)
y = np.round((y/(s[0]*s[1]))*(255)) 
```

```python
for i in range(s[0]):
    for j in range(s[1]):
        k = imgGray[i,j]
        imgGray[i,j] = y[k]
```

```python
histe = Hist(imgGray)
x = histg.reshape(1,256)
y=np.array([])
y = np.append(y,x[0,0])
print("The shape of my histogram",histg.shape)
```

```python
plt.figure()
plt.plot(histe)
plt.savefig('Equalized Histogram')
plt.show()
```
